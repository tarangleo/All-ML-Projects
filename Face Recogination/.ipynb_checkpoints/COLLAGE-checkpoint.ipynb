```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img_1=cv2.imread('top_left.jpg')

img1=cv2.cvtColor(img_1,cv2.COLOR_BGR2RGB)

print(img_1.shape)
```

```python
plt.imshow(img1)
plt.show()
```

```python
img_2=cv2.imread('top_right.jpg')
img2=cv2.cvtColor(img_2,cv2.COLOR_BGR2RGB)

img_3=cv2.imread('bottom_left.jpg')
img3=cv2.cvtColor(img_3,cv2.COLOR_BGR2RGB)

img_4=cv2.imread('bottom_right.jpg')
img4=cv2.cvtColor(img_4,cv2.COLOR_BGR2RGB)

img_5=cv2.imread('center.jpeg')
img5=cv2.cvtColor(img_5,cv2.COLOR_BGR2RGB)
```

```python
img1=cv2.resize(img1,(200,200))
img2=cv2.resize(img2,(200,200))
img3=cv2.resize(img3,(200,200))
img4=cv2.resize(img4,(200,200))
img5=cv2.resize(img5,(100,100))
```

```python
img1=cv2.copyMakeBorder(img1,10,0,10,10,cv2.BORDER_ISOLATED,value=(0,0,0))
img2=cv2.copyMakeBorder(img2,10,0,0,10,cv2.BORDER_CONSTANT,value=(0,0,0))
img3=cv2.copyMakeBorder(img3,10,10,10,10,cv2.BORDER_CONSTANT,value=(0,0,0))
img4=cv2.copyMakeBorder(img4,10,10,0,10,cv2.BORDER_CONSTANT,value=(0,0,0))
img5=cv2.copyMakeBorder(img5,10,10,10,10,cv2.BORDER_CONSTANT,value=(0,0,0))
```

```python
plt.imshow(img1)

```
