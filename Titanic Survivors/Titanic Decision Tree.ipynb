## Decision Tree Prepairation

```python
def entropy(col):
    a=np.unique(col,return_counts=True)
    # print(col.shape[0])
    entropy=0
    for i in a[1]:
        p=float(i)/col.shape[0]
        entropy+=(-1*p*np.log2(p))
    
    # print(entropy)
    return entropy
```

```python
def divide(x_cols,fkey,fval):
    left=pd.DataFrame([],columns=x_cols.columns)
    right=pd.DataFrame([],columns=x_cols.columns)
    for i in range(x_cols.shape[0]):
        val=x_cols[fkey].loc[i]
        
        if val<fval:
            left=left.append(x_cols.loc[i])
        else:
            right=right.append(x_cols.loc[i])
    return left,right
```

```python
def info_gain(x_cols,fkey,fval):
    left,right=divide(x_cols,fkey,fval)
    l,r=float(left.shape[0])/x_cols.shape[0],float(right.shape[0])/x_cols.shape[0]
    # print(entropy(right.survived))
    if left.shape[0]==0 or right.shape[0]==0:
        return -100000
    ig=entropy(x_cols.survived)-l*entropy(left.survived)-r*entropy(right.survived)
    return ig
```

## Decision Tree

```python
list(data.columns)
```

```python
class Decision_Tree():
    def __init__(self,depth=0,max_depth=5):
        self.depth=depth
        self.max_depth=max_depth
        self.target=None
        self.left=None
        self.right=None
        self.fkey=None
        self.fval=None
    def train(self,xtrain):
        features=list(xtrain.columns)
        # print(features)
        features.remove("survived")
        information_gain=[]
        
        for i in features:
            ig=info_gain(xtrain,i,xtrain[i].mean())
            information_gain.append(ig)
        
        self.fkey=features[np.argmax(information_gain)]
        self.fval=xtrain[self.fkey].mean()
        
        print("feature Tree of:",self.fkey)
        
        left_tree,right_tree=divide(xtrain,self.fkey,self.fval)
        
        left_tree=left_tree.reset_index(drop=True)
        right_tree=right_tree.reset_index(drop=True)
        
        if left_tree.shape[0]==0 or right_tree.shape[0]==0:
            if xtrain.survived.mean()>=0.5:
                self.target= "Survived"
            else:
                self.target= "Dead"
            return
        if self.depth==self.max_depth:
            if xtrain.survived.mean()>=0.5:
                self.target= "Survived"
            else:
                self.target= "Dead"
            return
        
        self.left=Decision_Tree(self.depth+1,self.max_depth)
        self.left.train(left_tree)
        
        self.right=Decision_Tree(self.depth+1,self.max_depth)
        self.right.train(right_tree)
        
        if xtrain.survived.mean()>0.5:
            self.target= "Survived"
        else:
            self.target= "Dead"
        return
    def predict(self,x_test):
        if x_test[self.fkey]>self.fval:
            if self.right==None:
                return self.target
            return self.right.predict(x_test)
        else:
            if self.left==None:
                return self.target
            return self.left.predict(x_test)


        
```

```python
dt=Decision_Tree()
dt.train(data)
```
