```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
        
```

```python
iris=load_iris()
x,y = iris.data[:,2:], iris.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,shuffle=True,random_state=123)
```

```python
print(f'X_train.shape: {x_train.shape}')
print(f'y_train.shape: {y_train.shape}')
print(f'X_test.shape: {x_test.shape}')
print(f'y_test.shape: {y_test.shape}')
```

```python
from mlxtend.plotting import scatterplotmatrix
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k_model = KNeighborsClassifier(n_neighbors=3)
k_model.fit(x_train,y_train);
```

```python
y_pred=k_model.predict(x_test)
c_p= (y_pred==y_test).sum()
accuracy=(c_p/y_test.shape[0])*100
```

```python
print (accuracy)
```

```python
import matplotlib.pyplot as plt
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x_train,y_train,k_model)
plt.xlabel("length")
plt.ylabel("width")
plt.legend(loc="upper left")
```

```python
plot_decision_regions(x_test,y_test,k_model)
plt.xlabel("length")
plt.ylabel("width")
plt.legend(loc="upper left")
```
